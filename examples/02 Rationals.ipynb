# Rational Examples

```python
# Initialize
import numpy as np
from src.piecewise.polynomial import Polynomial
from src.piecewise.rational import Rational
import matplotlib.pylab as plt
```

## Example 1: Basic usage

```python
# First rational function: (x² + 1) / (x - 1)
r1 = Rational(
    numerator=Polynomial([1, 0, 1], name="x²+1"),
    denominator=Polynomial([1, -5], name="x-5"),
    name="r1"
)

# Second rational function: x / (x² + 1)
r2 = Rational(
    numerator=Polynomial([1, 0], name="x"),
    denominator=Polynomial([1, 0, 1], name="x²+1"),
    name="r2"
)

print(f"r1 = {r1}")
print(f"r2 = {r2}")
print(f"r1(2) = {r1(2):.3f}")
print(f"r2(1) = {r2(1):.3f}")

# Graphs
x1 = np.linspace(-3, 3, 1000)

plt.plot(x1, r1(x1), 'b-', label='r1(x) = (x²+1)/(x-5)', linewidth=2)
plt.plot(x1, r2(x1), 'r-', label='r2(x) = x/(x²+1)', linewidth=2)
plt.ylim(-10, 10)
plt.title('Basic rational functions')
plt.legend()
plt.grid(True, alpha=0.3)
```

## Example 2: Numerical operations

```python
r3 = r1 + r2
r4 = r1 * r2
r5 = r1 - r2

print(f"r1 + r2 = {r3}")
print(f"r1 * r2 = {r4}")
print(f"r1 - r2 = {r5}")

x2 = np.linspace(-2, 2, 500)

plt.plot(x2, r1(x2), 'b-', label='r1(x)', linewidth=2)
plt.plot(x2, r2(x2), 'r-', label='r2(x)', linewidth=2)
plt.plot(x2, r3(x2), 'g--', label='r1 + r2', linewidth=2)
plt.title('Addition of rational functions')
plt.legend()
plt.grid(True, alpha=0.3)
```

## Example 3: Derivatives

```python
r1_prime = r1.diff(1, name="r1'")
r1_double_prime = r1.diff(2, name="r1''")

print(f"r1 = {r1}")
print(f"r1' = {r1_prime}")
print(f"r1'' = {r1_double_prime}")

x3 = np.linspace(-2, 2, 500)

plt.plot(x3, r1(x3), 'b-', label='r1(x)', linewidth=2)
plt.plot(x3, r1_prime(x3), 'r-', label="r1'(x)", linewidth=2)
plt.plot(x3, r1_double_prime(x3), 'g-', label="r1''(x)", linewidth=2)
plt.title('Derivatives of rational function')
plt.legend()
plt.grid(True, alpha=0.3)
```

## Example 4: Scaling and Shifting

```python
r_shifted = r2.shift(1, name="r2(x+1)")
r_scaled = r2.scale(2, name="r2(2x)")

print(f"r2 = {r2}")
print(f"r2(x+1) = {r_shifted}")
print(f"r2(2x) = {r_scaled}")

x4 = np.linspace(-3, 3, 500)
plt.plot(x4, r2(x4), 'b-', label='r2(x) = x/(x²+1)', linewidth=2)
plt.plot(x4, r_shifted(x4), 'r--', label='r2(x+1)', linewidth=2)
plt.plot(x4, r_scaled(x4), 'g--', label='r2(2x)', linewidth=2)
plt.title('Scaling and shifting')
plt.legend()
plt.grid(True, alpha=0.3)
```

## Example 5: Inversion and power

```python
r_inv = r2.updown()
r_power = r2 ** 2
r_division = r1 / r2

print(f"r2 = {r2}")
print(f"1/r2 = {r_inv}")
print(f"r2² = {r_power}")
print(f"r1 / r2 = {r_division}")

x5 = np.linspace(-2, 2, 500)

plt.plot(x5, r2(x5), 'b-', label='r2(x)', linewidth=2)
plt.plot(x5, r_inv(x5), 'r-', label='1/r2(x)', linewidth=2)
plt.plot(x5, r_power(x5), 'g--', label='r2(x)²', linewidth=2)
plt.ylim(-3, 3)
plt.title('Invesrion and powers')
plt.legend()
plt.grid(True, alpha=0.3)
```

## Example 6: Integration

```python
try:
    integral, error = r2.integral(0, 1)
    print(f"∫₀¹ r2(x) dx = {integral:.6f} (error: {error:.2e})")
except Exception as e:
    print(f"Integration failed: {e}")
```
