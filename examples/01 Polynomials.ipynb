# Polynomial Examples

```python
# Initialize
import numpy as np
from src.piecewise.polynomial import Polynomial
import matplotlib.pylab as plt
```

## Example 1: Basic operations

```python
p1 = Polynomial([1, 2, 1], name="p1")  # x² + 2x + 1
p2 = Polynomial([1, -1], name="p2")    # x - 1

print(f"p1 = {p1}") # Simple print
print(repr(p1))     # Advanced print
print(f"p2 = {p2}")
print(f"p1 + p2 = {p1 + p2}")
print(f"p1 - p2 = {p1 - p2}")
print(f"p1 * p2 = {p1 * p2}")
print(f"p1 * 2 = {p1 * 2}")
print(f"p1(3) = {p1(3)}")

x = np.linspace(-2, 2, 100)
plt.plot(x, p1(x), 'b-', label='p1(x) = x² + 2x + 1')
plt.plot(x, p2(x), 'r-', label='p2(x) = x - 1')
plt.plot(x, (p1 + p2)(x), 'g--', label='p1 + p2')
plt.title('Basic operations')
plt.legend()
plt.grid(True)
```

## Example 2: Derivatives

```python
p3 = Polynomial([1, 0, -2, 1], name="p3")  # x³ - 2x + 1
p3_deriv = p3.diff(1, name="p3'")         # 3x² - 2
p3_deriv2 = p3.diff(2, name="p3''")       # 6x

print(f"p3 = {p3}")
print(f"p3' = {p3_deriv}")
print(f"p3'' = {p3_deriv2}")

plt.plot(x, p3(x), 'b-', label='p3(x) = x³ - 2x + 1')
plt.plot(x, p3_deriv(x), 'r-', label="p3'(x) = 3x² - 2")
plt.plot(x, p3_deriv2(x), 'g-', label="p3''(x) = 6x")
plt.title('Polynomial derivatives')
plt.legend()
plt.grid(True)
```

## Example 3: Shifting

```python
p4 = Polynomial([1, 0, 0], name="p4")  # x²
p4_shifted = p4.shift(1, name="p4(x+1)")  # (x+1)² = x² + 2x + 1

print(f"p4 = {p4}")
print(f"p4(x+1) = {p4_shifted}")

plt.plot(x, p4(x), 'b-', label='p4(x) = x²')
plt.plot(x, p4_shifted(x), 'r--', label='p4(x+1) = (x+1)²')
plt.title('Polynomial shifting')
plt.legend()
plt.grid(True)
```

## Example 4: Scaling

```python
p5 = Polynomial([1, 0, 0], name="p5")  # x²
p5_scaled = p5.scale(2, name="p5(2x)")  # (2x)² = 4x²

print(f"p5 = {p5}")
print(f"p5(2x) = {p5_scaled}")

plt.plot(x, p5(x), 'b-', label='p5(x) = x²')
plt.plot(x, p5_scaled(x), 'r--', label='p5(2x) = 4x²')
plt.title('Ploynomial scaling')
plt.legend()
plt.grid(True)
```

## Example 5: Combined operations

```python
# (x² + 1) * (x - 1) + 2x
p6 = (Polynomial([1, 0, 1]) * Polynomial([1, -1])) + Polynomial([2, 0])

print(f"p6 = {p6}")
print(f"derivative = {p6.diff()}")

plt.plot(x, p6(x), 'b-', label='(x²+1)(x-1) + 2x')
plt.plot(x, p6.diff()(x), 'r--', label='Derivative')
plt.title('Combined operations')
plt.legend()
plt.grid(True)
```

## Example 6: Copyable functionality

```python
p7 = Polynomial([1, 1], name="Prototype")  # x + 1
p7_copy = p7.deepcopy()
p7_copy.name = "Copy"

print(f"Prototype: {p7}")
print(f"Copy: {p7_copy}")
print(f"Same object: {p7 is p7_copy}")
print(f"Same coeffs: {np.array_equal(p7.coeffs, p7_copy.coeffs)}")

plt.plot(x, p7(x), 'b-', label='Prototype: x + 1')
plt.plot(x, p7_copy(x), 'ro', markersize=2, label='Copy')
plt.title('Copyable functionality')
plt.legend()
plt.grid(True)
```
