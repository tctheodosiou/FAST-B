# Points Examples

```python
# Initialize
import numpy as np
import matplotlib.pyplot as plt
from src.nurbs.point import Point
```

## Example 1: Basic usage

```python
# Create 2D point
p1 = Point([1.0, 2.0], name="Point in 2D")
print(f"p1 = {p1}")
print(f"p1.x = {p1.x}, p1.y = {p1.y}, p1.z = {p1.z}")

# Create 3D point
p2 = Point([2.0, 3.0, 1.0], name="Point in 3D")
print(f"p2 = {p2}")
print(f"p2[0] = {p2[0]}, p2[1] = {p2[1]}, p2[2] = {p2[2]}")

# Change coordinates
p1.x = 3.0
p1.y = 4.0
print(f"After coordinates change: p1 = {p1}")

p1.plot('ro', markersize=10, label='p1')
p2.plot('bo', markersize=10, label='p2')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Basic usage of Point')
plt.legend()
plt.grid(True, alpha=0.3)
plt.axis('equal')
```

## Example 2: Point calculations

```python
p3 = Point([1, 1])
p4 = Point([2, 3])

# Addition
p_sum = p3 + p4
print(f"p3 + p4 = {p_sum}")

# Subtraction
p_diff = p4 - p3
print(f"p4 - p3 = {p_diff}")

# Multiplications with scalar
p_scaled = 2 * p3
print(f"2 * p3 = {p_scaled}")

# Negation
p_neg = -p3
print(f"-p3 = {p_neg}")
```

## Example 3: Transformation matrix

```python
# Rotate 90 degrees
rotation_90 = np.array([[0, -1, 0],
                        [1, 0, 0],
                        [0, 0, 1]])

# Scaling
scaling = np.array([[2, 0, 0],
                    [0, 1.5, 0],
                    [0, 0, 1]])

p_original = Point([1, 1])
p_rotated = rotation_90 * p_original
p_scaled_matrix = scaling * p_original

print(f"Original: {p_original}")
print(f"After rotation 90°: {p_rotated}")
print(f"After scaling: {p_scaled_matrix}")
```

## Example 4: Mixed calculations

```python
# Parse from scalar
p_from_num = Point.parse(5)
print(f"Point.parse(5) = {p_from_num}")

# Mixed calculations
p_base = Point([2, 3])
p_mixed = p_base + 3  # Add scalar
print(f"p_base + 3 = {p_mixed}")

p_mixed2 = 2 * p_base - 1
print(f"2 * p_base - 1 = {p_mixed2}")
```

## Example 5: Geometrical shapes

```python
# Triangle
triangle_points = [
    Point([0, 0], name="A"),
    Point([2, 0], name="B"), 
    Point([1, 2], name="C"),
    Point([0, 0], name="A")  # Close shape
]

# Square
square_points = [
    Point([3, 0], name="D"),
    Point([5, 0], name="E"),
    Point([5, 2], name="F"),
    Point([3, 2], name="G"),
    Point([3, 0], name="D")  # Close shape
]

# Plot Triangle
for i in range(len(triangle_points)-1):
    p1, p2 = triangle_points[i], triangle_points[i+1]
    plt.plot([p1.x, p2.x], [p1.y, p2.y], 'b-', linewidth=2)

# Plot Square
for i in range(len(square_points)-1):
    p1, p2 = square_points[i], square_points[i+1]
    plt.plot([p1.x, p2.x], [p1.y, p2.y], 'r-', linewidth=2)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Geometrical shapes')
plt.grid(True, alpha=0.3)
plt.axis('equal')

plt.tight_layout()
plt.show()
```

## Example 6: 3D Points

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 3D σημεία
points_3d = [
    Point([0, 0, 0], name="origin"),
    Point([1, 0, 0], name="x_axis"),
    Point([0, 1, 0], name="y_axis"),
    Point([0, 0, 1], name="z_axis"),
    Point([1, 1, 1], name="diagonal")
]

colors = ['black', 'red', 'green', 'blue', 'purple']
labels = ['Origin', 'X-axis', 'Y-axis', 'Z-axis', 'Diagonal']

for i, point in enumerate(points_3d):
    ax.scatter(point.x, point.y, point.z, 
               c=colors[i], s=100, label=labels[i], alpha=0.7)

# Connections
ax.plot([0, 1], [0, 0], [0, 0], 'r-', linewidth=2, alpha=0.5)  # X-axis
ax.plot([0, 0], [0, 1], [0, 0], 'g-', linewidth=2, alpha=0.5)  # Y-axis  
ax.plot([0, 0], [0, 0], [0, 1], 'b-', linewidth=2, alpha=0.5)  # Z-axis

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Points')
ax.legend()
```
