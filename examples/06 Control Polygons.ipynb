# Control Polygon Examples

```python
# Initialize
import numpy as np
import matplotlib.pyplot as plt
from src.containers import Knotvector
from src.nurbs.cpoint import ControlPoint
from src.nurbs.cpolygon import ControlPolygon
```

## Example 1: Basic usage

```python
# Create control points
cp1 = ControlPoint([0, 0], w=1.0, name="P0")
cp2 = ControlPoint([1, 2], w=1.0, name="P1") 
cp3 = ControlPoint([3, 1], w=1.0, name="P2")
cp4 = ControlPoint([4, 0], w=1.0, name="P3")

# Create control polygon
polygon1 = ControlPolygon(ctrl_points=[cp1, cp2, cp3, cp4], name="Basic polygon")
print(f"Control Polygon: {polygon1}")
print(f"Number of points: {polygon1.nr_items}")
print(f"Weights: {polygon1.weights}")

polygon1.plot('ro-', linewidth=2, markersize=8, label='Control Polygon')
for i, key in enumerate(polygon1.keys):
    point = polygon1[key]
    plt.text(point.x + 0.1, point.y + 0.1, f'P{key}', fontsize=10)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Basic Control Polygon')
plt.legend()
plt.grid(True, alpha=0.3)
plt.axis('equal')
```

## Example 2: Control polygon with various weights

```python
# Create control points
cp_w1 = ControlPoint([0, 0], w=1.0, name="W1")
cp_w2 = ControlPoint([2, 3], w=2.0, name="W2")  # larger weight
cp_w3 = ControlPoint([4, 1], w=0.5, name="W3")  # lower βάρος  
cp_w4 = ControlPoint([6, 0], w=1.0, name="W4")

polygon2 = ControlPolygon(ctrl_points=[cp_w1, cp_w2, cp_w3, cp_w4], name="weighted polygon")
print(f"Weights: {polygon2.weights}")

polygon2.plot('go-', linewidth=2, markersize=8, label='Weighted Polygon')

# Show weights
for i, key in enumerate(polygon2.keys):
    point = polygon2[key]
    plt.text(point.x + 0.1, point.y + 0.1, f'w={point.w}', fontsize=9)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Control Polygon with various weights')
plt.legend()
plt.grid(True, alpha=0.3)
plt.axis('equal')
```
